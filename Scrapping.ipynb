```python
%matplotlib inline

import requests
import pandas as pd
from bs4 import BeautifulSoup
import urllib.request
from io import BytesIO
import numpy as np
# from matplotlib import pyplot as plt

import base64
from PIL import Image
from IPython.display import HTML

def image_base64(im):
    if isinstance(im, str):
        im = get_thumbnail(im)
    with BytesIO() as buffer:
        im.save(buffer, 'jpeg')
        return base64.b64encode(buffer.getvalue()).decode()
    
def image_formatter(im):
    return f'<img src="data:image/jpeg;base64,{image_base64(im)}">'

base_url = 'https://www.tatacliq.com/electronics-mobile-phones/c-msh1210/page-1?q=%3Arelevance%3AinStockFlag%3Atrue&isFacet=true&facetValue=true'
r = requests.get(base_url)
soup = BeautifulSoup(r.text, "html.parser")
data = soup.find_all('li', class_="product-item")
product_name = []
price = []
discount = []
image = []
for item in data:
    name = item.find_all("a")[0]
    product_name.append(name.get("title"))
    prices = item.find_all("p",{"class":"sale"})[0]
    price.append(prices.text.strip())
    discounts = item.find_all("p",{"class":"savings"})    
    if not discounts:
        discount.append("NA")
    else:
        discounts=str(discounts[0].text.strip())
        discount.append(discounts.strip('( )').lstrip('-'))
    preview = item.find_all("img",class_="picZoomer-pic lazy")[0]
    img_url = preview.get("data-original")
    image.append(str('http:')+img_url)

image_data=pd.Series()

for url in image:
    resp = Image.open(urllib.request.urlopen(url))
    resp = resp.resize((100,150))
    image_data = image_data.append(pd.Series(resp), ignore_index=True)

pd.set_option('display.max_colwidth', -1)

df=pd.DataFrame({'Image':pd.Series(image_data),'Name' : pd.Series(product_name),'Price' : pd.Series(price),'Discount': pd.Series(discount)})
df.index = np.arange(1, len(df)+1)
HTML(df[['Image','Name','Price','Discount']].to_html(formatters={'Image':  image_formatter}, escape=False))
```
